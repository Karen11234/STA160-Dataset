```python
# load dependencies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
health_data = pd.read_csv('STA 160 Health Data Final.csv')
```

```python
health_data_risk_factors = health_data[["pct_diabetes", "pct_obesity", "pct_phys_inactivity", "pct_current_smoker"]]
health_data_soc_econ_envir_data = health_data[["pct_food_stmp", "income", "gini", "pct_poverty", "pct_unemployment",
                                    "air_quality",	"pct_park_access", "urban_rural_status"]]
health_data_health_care = health_data[["num_hospitals", "num_hospitals_intensive_care",	"num_hospitals_rehab", "num_hospitals_emergency",
                                       "num_hospitals_neuro", "pop_per_primary_dr",	"pop_per_cvd_dr", "pop_per_neuro_dr",
                                       "pop_per_neuro_surg_dr",	"pct_no_health_insur"]]
```

```python
# set figure size
plt.figure(figsize=(6,5))

# plot correlation matrix
sns.heatmap(health_data_risk_factors.corr(), annot=True)

# set the title
plt.title('Correlation Matrix: Risk Factors')
```

```python
# set figure size
plt.figure(figsize=(6,5))

# plot correlation matrix
sns.heatmap(health_data_soc_econ_envir_data.corr(), annot=True)

# set the title
plt.title('Correlation Matrix: Social, Economic, Environmental Data')
```

```python
# set figure size
plt.figure(figsize=(6,5))

# plot correlation matrix
sns.heatmap(health_data_health_care.corr(), annot=True)

# set the title
plt.title('Correlation Matrix: Health Care')
```
